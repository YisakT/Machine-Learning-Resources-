## Matplotlib exampe workflow

```python
# 0. immport matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data 
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plots
fig, ax = plt.subplots(figsize=(10,10)) # (width,height)

# 3. Plot the data
ax.plot(x,y)


# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save the whole figure

fig.savefig("images/sample-plot.png")
```

```python
# subplot option 1
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,
                                                   ncols=2,
                                                   figsize=(10,5))

# plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.random(1000));
```

```python
# subplots option 2

fig, ax=plt.subplots(nrows=2,
                    ncols=2,
                    figsize=(10,5))

#Plot to each different index

ax[0,0].plot(x, x/2);

ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
car_sales["Price"] = car_sales["Price"].astype(str)
```

```python
# using regx we are replacing $ , . from price with a empty space
# Regular Expression Mode: By default, str.replace()
# in pandas treats the pattern as a regular expression.
# Ensure that this behavior hasn't been altered. If needed,
# explicitly set the regex parameter to True:

car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '', regex=True)
car_sales
```

```python
# Remove last two zeros
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
# add a car sales date colume to dataframe
# pd.date_range
# This is a call to the date_range function from the pandas library 

car_sales["Sales Date"] = pd.date_range("1/1/2023", periods=len(car_sales))
car_sales
```

```python
# create a Total Sales colume
# we are using to cumsum to add the current row with the privious to get total 
# we have to chave it into int using astype(int)
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
# lets plot the total sales 

car_sales.plot(x='Sales Date', y ="Total Sales");
```

### Which one should you use? (pyplot vs matplotlib OO method?

* when ploting somting quickly its ok to use pyplot method
* when ploting someting advanced use more advanced OO method


```python
heart_disease.head()
```

```python
# filter the data from the table
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# OO method mixed with Pyplot method
fix,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);
# adjust the axis limit 
ax.set_xlim([45,100]);
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data

scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);


# customize

ax.set(title="Heart Disease and Cholestrol levels",
      xlabel="age",
      ylabel="Cholestrol")

# Add a legend which is a key for the table 

ax.legend(*scatter.legend_elements(), title="Target");

# add a horizontal line for the mean value

ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

```python
# create Subplot of chol,age,thalach 

fig, (ax0,ax1)= plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10))

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize ax0

ax0.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# Add legend to ax0

ax0.legend(*scatter.legend_elements(), title ="Target")

# Add a meanline

ax0.axhline(y=over_50["chol"].mean(),
           linestyle='--');

# Add data to ax1

scatter= ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"])

# Customize ax1

ax1.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Max Heart Rate");

#Add a legend to ax1

ax1.legend(*scatter.legend_elements(), title="Target");

# add a mean line

ax1.axhline(y=over_50["thalach"].mean(),
           linestyle='--');

# Add a title to the figure(overall plot)

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# customize your plot further

# set the style

plt.style.use('seaborn-v0_8-whitegrid')


# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data

scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="summer"); # this changes the color scheme 


# customize

ax.set(title="Heart Disease and Cholestrol levels",
      xlabel="age",
      ylabel="Cholestrol")

# Add a legend which is a key for the table 

ax.legend(*scatter.legend_elements(), title="Target");

# add a horizontal line for the mean value

ax.axhline(over_50["chol"].mean(),
           linestyle='--');
```

```python
# Customizing the y and x axis limitation
fig, (ax0,ax1)= plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10))

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter")

# Customize ax0

ax0.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholesterol")

# customize ax0 change x axis limits

ax0.set_xlim([50,80])


# Add legend to ax0

ax0.legend(*scatter.legend_elements(), title ="Target")

# Add a meanline

ax0.axhline(y=over_50["chol"].mean(),
           linestyle='--');

# Add data to ax1

scatter= ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="summer")

# Customize ax1

ax1.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Max Heart Rate");

# customize ax1 change x axis limits

ax1.set_xlim([50,80])


# customize ax1 change y axis limits

ax1.set_ylim([60,200])


#Add a legend to ax1

ax1.legend(*scatter.legend_elements(), title="Target");

# add a mean line

ax1.axhline(y=over_50["thalach"].mean(),
           linestyle='--');

# Add a title to the figure(overall plot)

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
